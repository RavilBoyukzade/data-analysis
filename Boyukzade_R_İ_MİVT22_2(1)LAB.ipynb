```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")


from sklearn.datasets import fetch_openml
name_of_dataset = "blood-transfusion-service-center"  # название датасета из таблицы (Боюкзаде Равиль)
data = fetch_openml(name=name_of_dataset)
df = pd.DataFrame(data.data, columns=data.feature_names)
# Добавление столбца "target" и заполнение его данными.
df['target'] = data.target
```
